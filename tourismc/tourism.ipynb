```python
# cleaning first name, last name columns

emp_proper = emp['first_name'].str.strip().str.title()
emp['first_name'] = emp_proper
emp_proper_last = emp['last_name'].str.strip().str.title()
emp['last_name'] = emp_proper_last

```

```python
# cleaning e-mail and position columns

email_prop = emp['e_mail'].str.strip().str.title()
emp['e_mail'] = email_prop
emp['position'] = emp['position'].str.replace('&', 'and')
pos_prop = emp['position'].str.strip().str.title()
emp['position'] = pos_prop
emp.head(25)
```

```python
# merging Bookings table with Tours

merged_bt = book.merge(tour, on='tour_id', how='inner')
merged_bt
```

```python
tour_no = merged_bt['tour_name'].value_counts()

plt.figure(figsize=(10, 6))
tour_no.plot(kind='bar')
plt.title('Number of bookings per tour')
plt.xlabel('Tour name')
plt.ylabel('Number of bookings')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Creating new column total_sale

merged_bt['total_sale'] = merged_bt['quantity'] * merged_bt['price_per_person']
merged_bt
```

```python
# Sales over years

years = merged_bt['booking_date'].dt.year
sales_over_years = merged_bt.groupby(years)['total_sale'].sum()
sales_over_years
```

```python
# Sales over months

months = merged_bt['booking_date'].dt.month
sales_over_months = merged_bt.groupby(months)['total_sale'].sum()
sales_over_months
```

```python
# Plotting the distribution of sales over years

plt.figure(figsize=(12, 6))
sales_over_years.plot(kind='line', marker='o', markersize=5)
plt.title('Distribution of sales over years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right', fontsize=8)  
plt.yticks(fontsize=8)  
plt.grid(True)
plt.show()
```

```python
# Histogram of total_sales

plt.figure(figsize=(8, 6))
plt.hist(merged_bt['total_sale'], bins=15, edgecolor='black')
plt.title('Distribution of total sales')
plt.xlabel('Payment amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Merging bt dataframe (bookings+tours) with employee

merged_bte = merged_bt.merge(emp, on='employee_id', how='left')
merged_bte.head(5)
```

```python
# Identifying top sales person

emps = merged_bte['first_name']
top_emp = merged_bt.groupby(emps)['total_sale'].sum()
top_emp_sorted = top_emp.sort_values(ascending=False)
top_emp_sorted
```

```python
#Plotting total_sales per employee

plt.figure(figsize=(10, 6))
top_emp_sorted.plot(kind='bar')
plt.title('Total sales per employee')
plt.xlabel('Employee name')
plt.ylabel('Sale amount')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
