```python
# Importing libraries

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading file

df = pd.read_csv(r"C:\Users\Gabon\Desktop\Nuttritional content of food dataset\Nutt.csv")
```

```python
# Inspecting dataset

df.info()
```

```python
# Inspecting all columns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
df=df.drop(['index'],axis=1)

df
```

```python
# Select  columns of interest

dfs = df[['Shrt_Desc', 'Energ_Kcal', 'Protein_(g)','Carbohydrt_(g)','Sugar_Tot_(g)', 'Calcium_(mg)', 'Lipid_Tot_(g)', 'Fiber_TD_(g)',
         'Sodium_(mg)','Potassium_(mg)','Magnesium_(mg)','Zinc_(mg)', 'Vit_B6_(mg)', 'Vit_B12_(µg)','Vit_D_µg',]]

```

```python
dfs.head()
```

```python
# Check for null values

dfs.isnull().sum()
```

```python
# Replacing null values with the average of each column

# Selecting only numeric values

numeric_columns = dfs.select_dtypes(include=['float64', 'int64']).columns
mean = dfs[numeric_columns].mean()

dfs = dfs.fillna(mean)

nan_count = dfs.isnull().sum()
nan_count

```

```python
# Renaming columns

dfs.rename(
    columns={"Shrt_Desc": "Food_item", "Energ_Kcal": "Energy_Kcal", "Carbohydrt_(g)": "Carbohydrates_(g)"},
    inplace=True,
)

dfs
```

```python
#Check for duplicate rows

dups = dfs.duplicated().sum()
dups
```

```python
# Finding top 30 food items with highest protein content.

grouped_df = dfs.groupby('Food_item')


max_by_category = grouped_df['Protein_(g)'].max()

max_sort = max_by_category.sort_values(ascending = False)

top_30 = top_protein = pd.DataFrame({'Food_item': max_sort.index, 'Total_Protein_(g)': max_sort.values})

top_30.head(30)
```

```python
# Visualizing top 30 proteic food items 

top_30 = top_protein.sort_values(by='Total_Protein_(g)', ascending=False).head(30)

plt.figure(figsize=(20, 8))
plt.bar(top_30['Food_item'], top_30['Total_Protein_(g)'])

plt.title('Top 30 Foods - Protein Distribution', fontsize=16)
plt.xlabel('Food Items', fontsize=13)
plt.ylabel('Total Protein (g)', fontsize=15)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=15)


plt.show()
```

```python
# Looking for top sugar items

grouped_df = dfs.groupby('Food_item')


max_by_category = grouped_df['Sugar_Tot_(g)'].max()

max_sort = max_by_category.sort_values(ascending = False)

top_30 = rdfs = pd.DataFrame({'Food_item': max_sort.index, 'Sugar_Tot_(g)': max_sort.values})

top_30.head(30)
```

```python
# Food items with highest sugar content

top_30 = rdfs.sort_values(by='Sugar_Tot_(g)', ascending=False).head(30)

plt.figure(figsize=(20, 8))
plt.bar(top_30['Food_item'], top_30['Sugar_Tot_(g)'])

plt.title('Top 30 Foods - Sugar Distribution', fontsize=16)
plt.xlabel('Food Items', fontsize=13)
plt.ylabel('Sugar in grams', fontsize=15)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=15)


plt.show()
```

```python
# Relationship between protein-carbohydrayes, sugar-kcal, protein-kcal

plt.figure(figsize=(10, 5))
sns.scatterplot(x='Protein_(g)', y='Carbohydrates_(g)', data=dfs, alpha=0.7, color='blue')

plt.title('Protein - Carbohydrates relationship')
plt.xlabel('Protein (g)')
plt.ylabel('Carbohydrates (g)')
plt.show()


plt.figure(figsize=(10, 5))
sns.scatterplot(x='Energy_Kcal', y='Sugar_Tot_(g)', data=dfs, alpha=0.7, color='green')


plt.title('Sugar - Kcal relationship')
plt.xlabel('Energy (Kcal)')
plt.ylabel('Sugar(g)')
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x='Energy_Kcal', y='Protein_(g)', data=dfs, alpha=0.7, color='red')

plt.title('Protein - Kcal relationship')
plt.xlabel('Energy (Kcal)')
plt.ylabel('Protein(g)')
plt.show()
```

```python
# Distribution of Kcal

sug_hist = dfs[['Energy_Kcal']]

plt.figure(figsize=(10, 5))

plt.hist(sug_hist['Energy_Kcal'], bins=15, color='purple', edgecolor='black')
plt.title('Kcal Distribution')
plt.xlabel('Energy_Kcal')
plt.ylabel('Frequency')
plt.show()
```
